#### Plotted the growth in terms of population for 5 counties over 15 years.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib.ticker import StrMethodFormatter
```

```python
pop = pd.read_excel("./world_population.xlsx")
pop.head()
```

```python
pop.info
```

```python
def plot_population():
    for ctry in pop.sort_values('Population Total', ascending = False).Country.unique()[2:7]:
        population = pop[pop.Country == ctry]['Population Total']
        yrs = pop[pop.Country == ctry].Year
        plt.plot(yrs,population, label = ctry )

    plt.title('World Population 2000 - 2015')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x}'))
    plt.legend()
    plt.style.use('fivethirtyeight')
plot_population()
```

The plot clearly shows the countries in descending order based on population over the 15 years. It also shows that relatively, Nigeria has shown the most growth in terms of population, while when it comes to Brazil, the increase seems to be much less since the slope of the plot is also smaller.
